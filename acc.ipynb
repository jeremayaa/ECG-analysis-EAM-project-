```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import signal
```

```python


acc = []

df = pd.read_csv("EAM Projekt/JP/acc_pompki.lvm", sep='\t', decimal=',')

for column in df.columns:
    acc.append(np.array(df[column]))

acc.pop(0)

step = 1

time = np.linspace(0, len(acc[0]), len(acc[0]))

pos = []
vel = []
for a in acc:
    mean_a = np.mean(a, axis=0)
    a = a - mean_a
    plt.plot(time, a)
    v = np.cumsum(a) * step

    vel.append(v)
    p = np.cumsum(v) * step
    pos.append(p-v)




plt.show()
```
